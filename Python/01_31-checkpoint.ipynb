# DB (SQLite)
- 데이터 저장 -> file , DB 
- SQLite : 데이터베이스 관리 시스템 (DBMS)

```python
import sqlite3

conn = sqlite3.connect("mydb.db")         # db 연결
cursor = conn.cursor()                    # cursor 만들기

# 테이블 만들기

sql = '''
    create table student (
        stdid int primary key,
        stdname varchar(30)
        )
'''

cursor.execute(sql)         # sql 문장 보내기
cursor.close()
conn.close()
```

```python
conn = sqlite3.connect("mydb.db")         # db 연결
cursor = conn.cursor()                    # cursor 만들기

# 데이터 인서트 하기

sql1 = "insert into student values(4, '강성빈_1')"
sql2 = "insert into student values(5, '백희연_1')"
sql3 = "insert into student values(6, '김수현_1')"

cursor.execute(sql1)       # sql 문장 보내기
cursor.execute(sql2)
cursor.execute(sql3)

conn.commit()

cursor.execute("select * from student")  # sql 문장 보내기

for row in cursor.fetchall():           # 전부 가져 와라
    print(row)

cursor.close()
conn.close()
```

```python
import cx_Oracle
```

```python
class Employee:
    def __init__(self, empid, fname, lname, salary):
        self.empid = empid
        self.fname = fname
        self.lname = lname
        self.salary = salary
    
    def empinfo(self):
        print("직원번호 : {}".format(self.empid))
        print("직원이름 : {}".format(self.fname))
        print("직원성 : {}".format(self.lname))
        print("직원급여 : {}".format(self.salary))
        print("-"*50)
```

```python
conn = cx_Oracle.connect("hr/hr@localhost:1521/xe")         # oracle 연결
cursor = conn.cursor()                    # cursor 만들기

cursor.execute("select * from employees where salary >= 15000")   # sql 문장 보내기

emplist = []

for row in cursor.fetchall():            # 전부 가져 와라
    
    emp = Employee(row[0], row[1], row[2], row[7])
    emplist.append(emp)    
    
cursor.close()
conn.close()
```

```python
for emp in emplist:
    emp.empinfo()
```

```python
# 연습문제 #
# departments table 의 모든 정보를 수집합니다.
# list에 담기 (class 형태)
# list에 담은 데이터 출력
```

```python
class Depart:
    def __init__(self, empid, name, deptid, salary):
        self.empid = empid
        self.name = name
        self.deptid = deptid
        self.salary = salary
    
    def deptinfo(self):
        print("직원번호 : {}".format(self.empid))
        print("직원이름 : {}".format(self.name))
        print("직원부서 : {}".format(self.deptid))
        print("직원급여 : {}".format(self.salary))
        print("-"*50)
```

```python
conn = cx_Oracle.connect("hr/hr@localhost:1521/xe")         # oracle 연결
cursor = conn.cursor()                    # cursor 만들기

cursor.execute("select * from departments")   # sql 문장 보내기

emplist = []

for row in cursor.fetchall():            # 전부 가져 와라

    dept = Depart(row[0], row[1], row[2], row[3])
    emplist.append(dept)    
    
cursor.close()
conn.close()
```

```python
for dept in emplist:
    dept.deptinfo()
```

# 웹 크롤링

- 프로토콜 : http, https (HyperText Transfer Protocol)
- 다양한 http(s) 클라이먼트
  - GUI 브라우저 : Chrome, Internet Explorer 등
  - CLI 브라우저 : w3m, elinks, lynx 등
- 파이썬 라이브러리/프레임워크 : requests(첫 응답만 받고 추가 요청이 없음), selenium(자동화 할 수 있음), Scrapy

- 요청 방식 
  1. get 방식 : 조회 목적
  2. post 방식 : 추가, 수정, 삭제 등의 요청시 사용 → CRUD (creat = insert, read = select, update, delete)

- Header
   1. http 요청/응답시에 사용된다.
   2. Key/Value 형식이다.
   3. User-Agent : 브라우저 종류
   4. Referrer : 이전 페이지 URL

```python
# 다음 실시간 검색어 순위 가져오기

from bs4 import BeautifulSoup

url = "https://www.daum.net/"
response = requests.get(url)
html = response.text

# DOM (Document Object Model)
soup = BeautifulSoup(html, "html.parser")
soup                            ## html 에 비해 더 정리되서 보여짐


# CSS Selector : #(id)  .(class)   >(자식(하위))   공백(자손)
# div 태그가 class 속성 값이 realtime_part 그리고 자식이 ol 그리고 자식이 li
search = "div.realtime_part > ol > li a[tabindex = '-1']"

for tag in soup.select(search):
    print(tag.text)
```
