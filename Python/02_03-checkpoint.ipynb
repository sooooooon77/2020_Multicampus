## 인터파크 해외 여행 상품 제목 크롤링

```python
# 1. 인터파크 사이트에서 투어 클릭
#      http://tour.interpark.com/?smid1=header&smid2=tour
# 2. 검색창에 여행지를 입력
# 3. 검색버튼 클릭
# 4. 더보기 버튼 클릭
```

```python
browser = webdriver.Chrome("driver/chromedriver.exe")
browser.get("http://tour.interpark.com/?smid1=header&smid2=tour") 

browser.find_element_by_css_selector("#SearchGNBText").clear()
browser.find_element_by_css_selector("#SearchGNBText").send_keys("유럽")
browser.find_element_by_css_selector("button.search-btn").click()

time.sleep(3)

browser.find_element_by_css_selector("button.moreBtn").click()
```

```python
# 5. 해외 여행 상품 제목 가져오기
```

```python
hlist = browser.find_elements_by_css_selector("h5.proTit")

tourlist = []

for title in hlist:
    #print(title.text)
    
    s = title.get_attribute("onclick")
    
    ## s에서 url만 가져오기
    idx = s.find("'")     ##작은 따옴표의 위치를 찾자.
    s = s[idx+1:]
    idx = s.find("'")
    url = s[:idx]
    #print(url)
    
    tourlist.append((title.text, url))
```

selenium 보다 더 간단한 방법 => 이 방법을 못쓰는 사이트가 있기도 함!!

```python
# 1. 인터파크 사이트에서 투어 클릭
# 2. 검색창에 여행지를 입력
# 3. 검색버튼 클릭
# 4. 더보기 버튼 클릭
# 5. F12  ->  Network  ->  XHR  ->  GetPaycoResultForCategory  ->  preview : 필요한 데이터가 다 들어있는지 확인
# 6. Header  ->  맨 위에 Request URL 을 url 로 가져오기 
# 7. Header  ->  맨 밑에 Form Data 을 param(파라미터) 로 가져오기 
# 8. param 에서 pageSize 는 데이터 개수 (얻고 싶은 데이터 수 만큼 숫자를 넣어라)


url ="http://search-tour.interpark.com/Payco/Search/GetPaycoResultForCategory"

header = {}
param = {
    "isSuggest": "true",
    "word": "유럽",
    "categorycode1": "R",
    "categorycode2": "",
    "offset": "1",
    "pageSize": "100",       
    "geotype": "",
    "filtersort": "relevance",
    "filterFlag": "false",
    "priceFilterFlag": "false"
}

response = requests.post(url, params = param)
jsonObj = response.json()
for goods in jsonObj['datalist']:
    print(goods["goodsname"])          # 상품명
    print(goods["price"])              # 가격
    print(goods["customervaluation"])  # 고객 평점
    print("-"*80)
```
